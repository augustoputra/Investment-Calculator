```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
!pip install tabulate
from tabulate import tabulate
```

```python
plt.ioff()
def investment_calc(current_wealth, money_invested, rate, years):
    wealth_withrate = []
    wealth_withnorate = []
    
    list_table = []
    
    for i in range(1, years+1):
        y = (1+rate/1200)**(12*i)
        future_val = int((current_wealth*y + money_invested*(    (y-1)/(rate/1200)  )))
        future_val_nointerets = int(current_wealth + money_invested*i*12)
        
        wealth_withrate.append(future_val)
        wealth_withnorate.append(future_val_nointerets)
        x = "Rp {:,.0f}".format(future_val)
        y = "Rp {:,.0f}".format(future_val_nointerets)
        data = [i, y, x]
        list_table.append(data)
        
    headers = ["year", "without investment", "with investment"]
    print(tabulate(list_table, headers))
    

    
  
    
    plt.plot(range(1, years+1),  wealth_withrate,  marker='.', linestyle = '-', linewidth = 1, color = 'blue', label='With Interest Rate')
    plt.plot(range(1, years+1),  wealth_withnorate, marker='.', linestyle = '-', linewidth = 1, color = 'red', label='Without Interest Rate')
    plt.fill_between(range(1, years+1), wealth_withrate, color = 'blue', alpha = 0.3)
    plt.fill_between(range(1, years+1), wealth_withnorate, color = 'red', alpha = 0.3)
    plt.title('Wealth Over Time')
    plt.xlabel('years')
    plt.ylabel(f'wealth in Rupiah')
    plt.grid(True)
    plt.show()
    plt.tight_layout()
    plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, _: 'Rp {:,.0f}'.format(y)))






```

```python

#RUN THE CODE BELOW TO START CALCULATE YOUR INVESTMENT
try:
    current_wealth = int(input("Enter total amount of current investment/savings : "))
    money_invested = int(input("Enter desired amount of monthly contribution : "))
    rate = int(input("Enter estimate annual gain (ex: 6, 10, 15, 20) : "))
    years = int(input("Enter desired holding period (years) (ex: 5, 10, 20, etc) : "))

except ValueError:
    raise ValueError("Type of data has to be an integer !")
    
investment_calc(current_wealth, money_invested, rate, years)
```
